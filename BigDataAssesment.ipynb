#URL of HTML Article

```python
url = 'https://mars.nasa.gov/news/9390/nasa-retires-mineral-mapping-instrument-on-mars-orbiter/'
```

#Requesting HTML Content of URL

```python
response = requests.get(url)
html = response.text
```

#Extracting Text Content from HTML

```python
soup = BeautifulSoup(html, 'html.parser')
text = ''
image_tags = soup.find_all('img')
for tag in soup.find_all(['p', 'h1', 'h2', 'h3', 'li']):
    if tag.name == 'li':
        text += '- ' + tag.text.strip() + '. '
    else:
        text += tag.text.strip() + '. '
```
